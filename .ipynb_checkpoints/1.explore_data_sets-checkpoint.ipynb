## First going to filter for just lung since this will generate much smaller workable files

> Choosing lung because **A.** it has a relatively good number of samples ~300 **B.** There was a study [Hefferon et al.](https://www.ncbi.nlm.nih.gov/pubmed/14993601) which showed larger numbers of TG repeats in the splice acceptor of exon 9 of CFTR resulted in reduced exon 9 splicing efficiency. 

```python
dt = pd.read_table(gzip.open(FILE_ROOT + 'GTEx_Data_20150112_RNAseq_RNASeQCv1.1.8_metrics.tsv.gz'))
dt_lung = dt[dt['Note'].str.contains('Lung')]
dt_lung
```

```python
lung_samples = dt_lung['Sample']
lung_samples.reset_index(drop=True,inplace=True)
lung_samples.is_unique
```

## Generate new files which just have lung samples 

```python
OUT_DIR = '/storage/dana/spliceSTR/expression/'

for f in files[3:]:
    
    in_f = gzip.open(FILE_ROOT + f, 'r')
    out_f = open(OUT_DIR + f.split('txt.gz')[0] + 'lung.txt','w') 

    header = in_f.readline().decode().split()
    lung_indices = [header.index(i) for i in set(lung_samples)]
    lung_indices.extend([0,1,2,4])
    lung_header = [header[i] for i in set(lung_indices)]
    out_f.write('\t'.join(lung_header) + '\n')

    while True:
        line = in_f.readline().decode().split()
        if line:
            lung_line = [line[i] for i in set(lung_indices)]
            out_f.write('\t'.join(lung_line) + '\n')
        else:
            break

    in_f.close()
    out_f.close()
```

## For each file type, plot a histogram of the first row 

```python
files = [f for f in listdir('/storage/dana/spliceSTR/expression/')]
dfs = {}
for f in files:
    f_name = f.split('.')[-3][2:]
    dfs[f_name] = pd.read_table(OUT_DIR + f)
    print(f_name)
    plt.hist(dfs[f_name].iloc[:,3:].iloc[0])
    plt.show()
```
