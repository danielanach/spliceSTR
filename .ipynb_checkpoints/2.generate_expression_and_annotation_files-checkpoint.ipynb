# MAKING EXON-centric annotation table

### These exon probe annotations were found on the (GTEx Portal data base)![https://www.gtexportal.org/home/datasets]

```python
exon_ann_df = pd.read_table('/storage/dana/spliceSTR/expression/gencode.v19.genes.patched_contigs_exons.txt')
exon_ann_df = exon_ann_df[['CHR','start','stop','Gene','strand']]
exon_ann_df.columns = ann_df.columns[5:]
exon_ann_df['probe.chr'] = exon_ann_df['probe.chr'].apply(lambda x:'chr' + str(x) )
exon_ann_df['gene.id'] = exon_ann_df['probe.id'].apply(lambda x:x.split('_')[0] )
exon_ann_df = pd.merge(exon_ann_df,ann_gene,on='gene.id') 
exon_ann_df = exon_ann_df[ann_df.columns]
exon_ann_df.sort_values(by=['gene.chr','gene.id','probe.id'])
exon_ann_df.reset_index(inplace=True,drop=True)
exon_ann_df.to_csv('/storage/dana/spliceSTR/expression/gencode.v19.annotations_exons.csv')
```

## Data necessary for TRANSCRIPT RATIO ANALYSIS

```python
transcript_df = pd.read_table(FILE_ROOT + 'GTEx_Data_20150112_RNAseq_Flux1.6_transcript_reads.lung.txt',
                              dtype={"Gene_Symbol": str,'TargetID':str},
                              index_col=0,
                              sep='\t',
                              skipinitialspace=True)
```

```python
transcript_df = transcript_df.sort_values('Gene_Symbol')
```

```python
transcript_df
```

## Generate the transcript ratio dataframe:
> #### Going gene by gene and taking all transcripts within that gene and assigning a ratio of expression for each patient

```python
transcript_symbols = list(transcript_df.index)
gene_symbols = list(transcript_df['Gene_Symbol'])
```

```python
new_transcript_df = pd.DataFrame(index= transcript_df.index,columns = transcript_df.columns)
```

```python
new_transcript_df['#_transcripts'] = 'hi'
```

```python
all_genes = transcript_df['Gene_Symbol']['ENST00000000233.5']
all_genes
```

```python
new_transcript_df
```

```python
gene_set = set(gene_symbols)
patients = transcript_df.columns[2:]

for gene in gene_set:
    
    gene_df = transcript_df[transcript_df['Gene_Symbol'] == gene]
    new_gene_df = pd.DataFrame(gene_df)
    num_transcripts = gene_df.shape[0]
    sum_patients = gene_df[patients].astype(float).sum()
    
    for patient in patients:
        for i in range(num_transcripts):
            if sum_patients[patient] > 0:
                new_gene_df[patient][i] = float(gene_df[patient][i])/sum_patients[patient]
            else:
                new_gene_df[patient][i] = 0
    #new_transcript_df[patients] = new_gene_df[patients]
    #ratio_array = pd.DataFrame(index = gene_df.index)
```
